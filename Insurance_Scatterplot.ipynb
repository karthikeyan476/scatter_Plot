```python
# Importing Pandas to read the file 
#  To download the file https://www.kaggle.com/mirichoi0218/insurance/home
import pandas as pd 
data = pd.read_csv("insurance.csv") 
data.head(11)
```

```python
#  Importing Seaborn library to visualize scatter plt.
import seaborn as sns
sns.scatterplot(x=data['bmi'], y=data['charges'])
```

```python
# Adding Regression line in the scatter plot.

sns.regplot(x=data['bmi'], y=data['charges'])
```

```python
# To understand how smoking affects the relationship between BMI and insurance costs,
# we can color-code the points by 'smoker', and plot the other two columns ('bmi', 'charges') on the axes.

sns.scatterplot(x=data['bmi'], y=data['charges'], hue=data['smoker'])
```

```python
# Implot is used to add two Regression line corresponding to smokers and nonsmokers

sns.lmplot(x="bmi", y="charges", hue="smoker", data=data)
```

```python
# This refer to this plot type as a categorical scatter plot, and it is  build with the sns.swarmplot command.
sns.swarmplot(x=data['smoker'],
              y=data['charges'])
```
